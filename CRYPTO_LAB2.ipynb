```python
#Python function to find the number of invertible elements which are less than let's say n, and relatively prime to n  using Euler’s Totient function

def prime(num):
    if num<2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num%i==0:
            return False
    return True

def totient(n):
    if prime(n):
        return n-1
    factors=[]
    t=n
    for i in range(2,int(n**0.5)+1):
        while t%i==0:
            factors.append(i)
            t//=i
    if t>1:
        factors.append(t)
    if len(factors) == 2:
        p, q = factors
        return (p - 1)*(q - 1)
    totient = n
    for prime_factor in set(factors):
        totient *= (1 - 1 / prime_factor)
    return int(totient)


n=int(input('Enter a number:'))
result = totient(n)
print("Phi(",n,")=",result)
```

```python
#Python function to implement Fermat’s Little Theorem to check the primality of an integer

def fermat(n, a):

  if n <= 1:
    return False
  if n == 2:
    return True
  if pow(a, n - 1, n) != 1:
    return False
  return True


num = 15
for a in range(2, num):
  if not fermat(num, a):
    print(num,"is not prime")
    break
else:
  print(num,"is likely prime")
```
